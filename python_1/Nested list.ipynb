 # Nested list :- A nested list in Python is a list that contains one or more other lists as its elements.

It contain list along with other list together


```python
data = ["sham","om",["d1","d2","d3"],"adi","sai","tejas"]
print(data)
```

```python
print(len(data))
```

```python
data.insert(4,"d4")
print(data)
```

```python
data[2].insert(1,"d6")
print(data)
```

```python
data[2].remove("hii")
print(data)
```

```python
data[2][1]="ruhi"
print(data)
```
