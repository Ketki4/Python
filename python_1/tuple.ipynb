# String : - String is stored in double quotes " "
# List : - List is stored in square bracket [   ]    are mutable(changable)
# Tuple : - Tuple is store in round bracket (   )     are immutable (unchangable)
# List (): - used to convert data into list



```python
D1 = "om","hari","adi","shiv"
print(D1)
```

```python
A = list(D1)
print(A)
```

```python
D1 = tuple(A)
print(D1)
```

```python
A[1]="data1"
print(A)
```

```python
D1 = tuple(A)
print(D1)
```
