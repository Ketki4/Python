```python
num_1 = int(input("enter your first number:"))

operation = str(input("enter the operation-------------->"))

num_2 = float(input("enter your second number:"))

print(num_1)

print(operation)

print(num_2)

if operation=="+":
    print("result",num_1+num_2)
elif operation =="-":
    print("result",num_1-num_2)
elif operation == "*":
    print("result",num_1*num_2)
else:
    print(none)
```
