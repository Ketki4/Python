```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"C:\Users\ketki\OneDrive\Desktop\excel\DV - 1 (1).xlsx")
```

```python
df
```

LINE CHART

```python

plt.figure(figsize=(10,6))
sns.lineplot(x="Date",y="Sales",data=df,color="r",marker="o")
plt.xticks(rotation=45)

plt.title("date wise sales")

plt.xlabel("")

plt.ylabel("")

plt.grid(color="grey")

plt.show()
```

BAR CHART

```python
sns.barplot(x="Region",y="Sales",data=df,errorbar=None,color="yellow",edgecolor="black",linewidth=2.5)

plt.title("region wise sales")

plt.show()
```

PIE CHART

```python
 d = df.groupby(["Region"])['Quantity'].sum()
```

```python
d= d.reset_index()
```

```python
plt.pie(d["Quantity"],labels=d["Region"],autopct="%.f%%")

plt.show()
```
