```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

LINE CHARTS

```python
data_1 = [2,4,6,8,10,12]

data_2 = [10,20,30,40,50,60]
```

```python
plt.plot(data_1,data_2,marker="o",color="g",linestyle = "--")

plt.xlabel("data_1",fontsize=14)

plt.ylabel("data_2",fontsize=14)

plt.title("Line Chart",fontsize=16)

plt.grid()

plt.show()
```

```python
Month_Name = ["Jan","Feb","March","April","May"]

Profit = [2000,5000,7000,9000,10000]
```

```python
plt.plot(Month_Name,Profit,marker="o",color="y",linestyle="--")

plt.xlabel("Month_Name",fontsize=14)

plt.ylabel("Profit",fontsize=14)

plt.title("Line Chart",fontsize=16)

plt.show()
```

BAR CHART

```python
Region = ["North","South","East","West"]

Profit = [2000,1400,750,600]
```

```python
plt.bar(Region,Profit,color="r",edgecolor="black",linewidth=2.5)

plt.xlabel("Region",fontsize=14)

plt.ylabel("Profit",fontsize=14)

plt.title("Bar chart",fontsize=16)

plt.show()
```

```python
City = ["Pune","Mumbai","Nagpur","Nashik"]

Profit = [2500,1400,670,550]
```

```python
plt.bar(City,Profit,color="g",edgecolor="black",linewidth=2.5)

plt.xlabel("City",fontsize=14)

plt.ylabel("Profit",fontsize=14)

plt.title("Bar Chart",fontsize=16)

plt.show()
```

PIE CHART

```python
Region = ["North","South","East","West"]

c  = ["red","pink","g","y"]

Sales = [1000,780,650,450]
```

```python
plt.pie(Sales,labels=Region,autopct="%.f%%",colors=c,startangle=110)

plt.title("Pie Chart",fontsize=16)

plt.show()
```

```python
Category = ["Books","Clothing","Food","Electronics"]

c  = ["lightblue","grey","yellow","orange"]

Profit  = [2300,1700,980,880]
```

```python
plt.pie(Profit,labels=Category,autopct="%.f%%",colors=c,startangle=120)

plt.title("Pie Chart",fontsize=16)

plt.show()
```
