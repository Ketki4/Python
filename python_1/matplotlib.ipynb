```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

LINE CHART

```python
data_1 = [2,3,4,5,6,7]

data_2 = [10,20,30,40,50,60]
```

```python
plt.plot(data_1,data_2,marker ="o",color="r",linestyle="--")

plt.grid()

plt.xlabel("data_1",fontsize=14)

plt.ylabel("data_2",fontsize=14)

plt.title("line chart",fontsize = 16)


plt.show()
```

BAR CHART

```python
region = ["west","south","north","east"]

profit = [20000,17000,14000,10000]
```

```python
plt.bar(region,profit,color="g",edgecolor="black",linewidth=2)

plt.title("bar chart",fontsize=15)

plt.xlabel("region",fontsize=14)

plt.ylabel("profit",fontsize=14)

plt.show()
```

PIE CHART

```python
region = ["west","south","north","east"]

c = ["green","skyblue","y","r"]

sales = [1500,1000,700,400]
```

```python
plt.pie(sales,labels=region,autopct="%.f%%",colors=c,startangle=110)

plt.title("pie chart",fontsize=16)

plt.show()
```
