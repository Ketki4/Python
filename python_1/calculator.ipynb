```python
Num1=float(input("Enter the first number:"))

operation = str(input("Enter operation--->(+,-,*,/)"))

Num2=float(input("Enter the second number:"))

print(Num1)

print(operation)

print(Num2)

if operation =="+":
    print(Num1+Num2)
elif operation =="-":
    print(Num1-Num2)
elif operation == "*":
    print(Num1*Num2)
elif operation =="/":
    print(Num1/Num2)
else:
    print("None")
```
