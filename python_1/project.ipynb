DATA VISALUZATION

```python
#['#8e44ad', '#a569bd', '#bb8fce', '#d2b4de','#ebdef0']

color = ['#a569bd','#bb8fce']

A1 = sns.countplot(x="Gender",data=df,palette=color,edgecolor="black")

plt.title("Gender Wise Count",family="Georgia",color="#3d0753",fontsize=14,fontweight="bold")

plt.xlabel("")

plt.ylabel("")

for i in A1.containers:
    A1.bar_label(i)
```

```python
#['#2ecc71', '#58d68d', '#82e0aa', '#abebc6','#d5f5e3']
d =['#2ecc71','#82e0aa']
plt.figure(figsize=(10,6))
A2 = sns.countplot(x="Age Group",data=df,hue="Gender",palette=d,edgecolor="black")

plt.title("Age Group wise Gender count",color="#075527",family="Georgia",fontsize=14,fontweight="bold")

plt.xlabel("")

plt.ylabel("")

for r in A2.containers:
    A2.bar_label(r)
```

```python
d = df.groupby(["Zone"])["Profit"].sum()
```

```python
d
```

```python
d = d.reset_index()
```

```python
d
```

```python
T = ['#8e44ad', '#a569bd', '#bb8fce', '#d2b4de','#ebdef0']

plt.pie(d["Profit"],labels=d["Zone"],autopct="%.f%%",startangle=295,colors=T)

plt.title("Zone wise Profit",color="#21022e",family="Georgia",fontsize=14,fontweight="bold")

plt.show()
```

```python
d1 = df.groupby("State",as_index=False)["Profit"].sum().sort_values(by="Profit",ascending=False).head(5)

K = ["#A14E05", "#d66e12", "#E58E42", "#D97A26","#cd5f2d"]

plt.Figure(figsize=(10,6))

plt.title("State wise Profit",color="#8C470C",family="Georgia",fontsize=14,fontweight="bold")


sns.barplot(x="State",y="Profit",data=d1,palette=K)
```
