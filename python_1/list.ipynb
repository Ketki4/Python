How to create a List


```python
a = ["ritu","gitu","sheenu"]
print(a)
```

# append # pop # insert # remove # extend

Append = It is used to add element in the last of the list

```python
a = ["ritu","gitu","sheenu"]
a.append("d3")
print(a)
```

Pop = It  is used to delete the last added element

```python
a.pop()
print(a)
```

Insert = It is used to add element at the specific position

```python
a.insert(1,"d4")
print(a)
    
```

Remove = It is used to remove if element is added double in the list

```python
a.remove("d4")
print(a)
```

Extend = It is used to add new elements that are multiple

```python
a.extend(["1,2,3"])
print(a)
```

How to reverse a string

```python
print(a[::-1])
```
