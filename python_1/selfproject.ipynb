```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ketki\Downloads\ecommerce sales.csv")
```

```python
color = ['#8e44ad', '#a569bd', '#bb8fce', '#d2b4de','#ebdef0']
plt.figure(figsize=(10,6))

A1 = sns.countplot(x="Category",data=df,palette=color,edgecolor="black")

plt.xticks(rotation=45)

plt.title("Category-wise Order Volume",fontsize=14,family="Georgia",fontweight="bold",color="#420b59")

plt.xlabel("")

plt.ylabel("")

plt.show()



for i in A1.containers:
    A1.bar_label(i)

```

```python
d = df.groupby("Product Name",as_index=False)["Profit"].sum().sort_values(by="Profit",ascending=False).head(5)

color = ["#0C7939", "#0AA64B", "#0BE369", "#40f48b","#81f2b2"]

plt.figure(figsize=(10,6))

plt.title("Top 5 Profitable Products",fontsize=14,family="Georgia",fontweight="black",color="#032511")

plt.xlabel("")

plt.ylabel("")


sns.barplot(x="Product Name",y="Profit",data=d,palette=color)



```

```python
d1 = df.groupby("Region")["Profit"].sum()
```

```python
d1
```

```python
d1 = d1.reset_index()
```

```python
d1
```

```python
o = ["#f08d36", "#9F500A", '#f0b27a', '#f5cba7','#fbe9e7']

plt.pie(d1["Profit"],labels=d1["Region"],autopct="%.f%%",startangle=200,colors=o)

plt.title("Region-Based Profit Insights",fontsize=14,family="Georgia",fontweight="bold",color="#482506")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.title("Sales Progress Over Dates",fontsize=14,family="Georgia",fontweight="bold")

plt.grid(color="black")


sns.lineplot(x="Order Date",y="Sales",data=d2,marker="o",color="r")
```
