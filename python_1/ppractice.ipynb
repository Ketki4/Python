```python
plt.figure(figsize=(10,6))

A1 = sns.countplot(x="Category",data=df,color="green",edgecolor="black",linewidth=2.3)

plt.xticks(rotation=45)

plt.title("Category wise order volume ",fontsize=14,fontweight="bold")

for i in A1.containers:
    A1.bar_label(i)

```
