```python

a = ['#8e44ad', '#a569bd', '#bb8fce', '#d2b4de','#ebdef0']
A1= sns.countplot(x="Gender",data=df,palette=a,edgecolor="black")

plt.title("Gender wise count",color="#5d2575",family="Georgia",fontsize=14,fontweight="bold")

for i in A1.containers:
    A1.bar_label(i)
```

```python
A2 = sns.countplot(x="Age Group",data=df,hue="Gender")
plt.figure(figsize=(10,6))
for r in A2.containers:
    A2.bar_label(r)
plt.show()
```
